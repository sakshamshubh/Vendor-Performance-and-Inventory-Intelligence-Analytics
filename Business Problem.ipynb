# Business Problem

Effective inventory and sales management are critical for optimizing profitability in the retail and wholesale industry.  
Companies need to ensure that they are not incurring losses due to inefficient pricing, poor inventory turnover, or vendor dependency. 
You have been given the **begin_inventory, end_inventory, purchase_price, purchases, sales, and vendor_invoice** tables, which will serve as the foundation for analyzing vendor performance, inventory turnover, pricing strategies, and overall profitability.


**Goals of this analysis:**
- Identify underperforming brands that require promotional or pricing adjustments.  
- Determine top vendors contributing to sales and gross profit.  
- Analyze the impact of bulk purchasing on unit costs.  
- Assess inventory turnover to reduce holding costs and improve efficiency.  
- Provide actionable recommendations for pricing, vendor management, and inventory strategies to ensure sustainable growth.
